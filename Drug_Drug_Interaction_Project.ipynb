```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Simulate a Clinical Dataset (Mock Data)
# In a real study, you would load this from a CSV: df = pd.read_csv('patients.csv')
data = {
    'Patient_ID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'Age': [65, 54, 72, 45, 60, 50, 78, 62, 55, 70],
    'Drug_A': ['Warfarin', 'Aspirin', 'Simvastatin', 'Metformin', 'Warfarin',
               'Amoxicillin', 'Digoxin', 'Lisinopril', 'Clarithromycin', 'Warfarin'],
    'Drug_B': ['Ciprofloxacin', 'Ibuprofen', 'Amlodipine', 'Sitagliptin', 'Aspirin',
               'Clavulanic Acid', 'Furosemide', 'Hydrochlorothiazide', 'Simvastatin', 'Rifampin'],
    'Interaction_Risk': ['High', 'Medium', 'Medium', 'None', 'High',
                         'None', 'Medium', 'None', 'High', 'High']
}

df = pd.DataFrame(data)

# 2. Data Analysis 
print("--- Clinical Data Summary ---")
print(df.head())

# Calculate prevalence of interactions
risk_counts = df['Interaction_Risk'].value_counts()
print("\n--- Interaction Risk Distribution ---")
print(risk_counts)

# Calculate percentage of High Risk patients
high_risk_pct = (len(df[df['Interaction_Risk'] == 'High']) / len(df)) * 100
print(f"\nAlert: {high_risk_pct}% of patients are at High Risk of DDI.")

# 3. Visualization 
plt.figure(figsize=(8, 5))
sns.barplot(x=risk_counts.index, y=risk_counts.values, palette='viridis')
plt.title('Distribution of Potential Drug-Drug Interactions (DDIs)')
plt.xlabel('Risk Severity')
plt.ylabel('Number of Patients')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save 
plt.savefig('ddi_analysis_plot.png')
print("\n[Success] Analysis complete. Chart saved as 'ddi_analysis_plot.png'")
```
